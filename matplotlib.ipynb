# MATPLOTLIB PYPLOT

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints,ypoints)
plt.show()
```

# MATPLOTLIB PLOTTING

```python
xpoints = np.array([1,2,3])
ypoints = np.array([3,5,1])

plt.plot(xpoints,ypoints)
plt.show()
```

## Plot only marker

```python
xpoints = np.array([1,2,3])
ypoints = np.array([3,5,1])
plt.plot(xpoints,ypoints,'o')
plt.show()
```

## Ploting for multiple points

```python
xpoints = np.array([2,7,4,9,12])
ypoints = np.array([4,9,1,2,3])

plt.plot(xpoints,ypoints)
plt.show()
```

## Default X points

```python
ypoints = np.array([5,2,7,1])

plt.plot(ypoints)
plt.show()

# If we give only one array of points,xpoints array will be default that is [0,1,2,3,..]
```

# MATPLOTLIB MARKERS

```python
xpoints = np.array([3,6,5,2])
ypoints = np.array([5,1,9,15])

plt.plot(xpoints,ypoints,marker='o')
plt.show()
```

```python
plt.plot(xpoints,ypoints,marker = '*')
plt.show()
```

## fmt Format Strings

```python
plt.plot(xpoints,ypoints,'o:r')
plt.show()
```

```python
plt.plot(xpoints,ypoints,'*--b')
plt.show()
```

```python
plt.plot(xpoints,ypoints,'d-.g')
plt.show()
```

## Marker Size

```python
#markersize or ms
plt.plot(xpoints,ypoints,'o-',ms=20)
plt.show()
```

```python
#markeredgecolor or mec

plt.plot(xpoints,ypoints,'o-',mec='r')
plt.show()
```

```python
#markerfacecolor or mfc

plt.plot(xpoints,ypoints,marker = 'o', mfc = 'r', ms = 20)
plt.show()
```

```python
#We can use hexadecimal values also
#To color entire marker,set both mec and mfc

plt.plot(xpoints,ypoints,marker = 'o',mec = '#aaaaaa', mfc = '#aaaaaa',ms=20)
plt.show()
```

# MATPLOTLIB LINE

# MATPLOTLIB LABELS AND TITLES

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)
plt.xlabel("Pulse")
plt.ylabel("Burnage")
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Pulse")
plt.ylabel("Burnage")
plt.title("Sports watch data")

plt.show()
```

## Set font properties

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = { "family" : "serif" , "color" : "blue", "size":15}
font2 = { "family" : "serif" , "color" : "brown", "size":15}
font3 = { "family" : "serif" , "color" : "red", "size":25}

plt.plot(x,y)
plt.xlabel("Pulse",fontdict=font1)
plt.ylabel("Burnage",fontdict=font2)
plt.title("Sports Watch Data",fontdict=font3)

plt.show()
```

```python
plt.title("Sports Watch Data",loc="left")
plt.xlabel("Pulse")
plt.xlabel("Burnage")
plt.plot(x,y)

plt.show()
```

# MATPLOTLIB SUBPLOT

## Display Multiple Plots

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,2,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 1)
plt.plot(x,y)

plt.show()
```

```python
#Subplot function takes three args, rows,column and index

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,1,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([10,5,7,2])

plt.subplot(2,3,1)
plt.plot(x,y)

x = np.array([0,1,2,3])
y = np.array([5,10,1,5])

plt.subplot(2,3,2)
plt.plot(x,y)

x = np.array([0,1,2,3])
y = np.array([10,15,5,0])

plt.subplot(2,3,3)
plt.plot(x,y)


x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)


plt.show()
```

## Colors

```python
#changing color for each scatter plots

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y,c='red')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x,y,c='#cf23bf')

plt.show()
```

```python
#Color on each point
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y,c=colors)
plt.show()
```

## ColorMap

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.arange(0,121,10)

plt.scatter(x,y,c=colors,cmap='viridis')
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.arange(0,121,10)

plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()

plt.show()
```

```python
y = np.random.randint(1000,size=1000)
x = np.random.randint(0,1000,size=1000)
colors = np.arange(1000)
plt.scatter(x,y,c=colors,cmap='winter')
plt.colorbar()
plt.show()
```

## Size

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y,s=50)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y,s=sizes)
plt.show()
```

## Alpha

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y,s=sizes,alpha=0.2)
plt.show()
```

```python
#Combine colormap and alpha and size
x = np.random.randint(1,100,size=100)
y = np.random.randint(1,100,size=100)
sizes = np.random.randint(10,1000,size=100)
colors = np.random.randint(0,101,size=100)

plt.scatter(x,y,s=sizes,c=colors,cmap='nipy_spectral',alpha=0.4)
plt.colorbar()
plt.show()
```

# MATPLOTLIB PIE CHARTS

```python
y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

## Labels


```python
y = np.array([35, 25, 25, 15])
labels = np.array(['Apples','Bananas','Cherries','Dates'])
plt.pie(y,labels=labels)
plt.show()
```

## Start Angle

```python
plt.pie(y,labels=labels,startangle=90)
plt.show()
```

## Explode

```python
myexp = np.array([0.2,0,0,0])
plt.pie(y,explode=myexp,labels=labels)
plt.show()
```

```python
plt.pie(y,explode=myexp,labels=labels,shadow=True)
plt.show()
```

```python
colors = np.array(['red','#4CAF50','black','b'])
plt.pie(y,explode=myexp,labels=labels,shadow=True,colors=colors)
plt.show()
```

## Legend

```python
plt.pie(y,explode=myexp,labels=labels)
plt.legend()
plt.show()
```

```python
plt.pie(y,explode=myexp,labels=labels)
plt.legend(title='Fruits')

plt.show()
```
